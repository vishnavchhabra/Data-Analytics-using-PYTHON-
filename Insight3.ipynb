```python
sales.dropna(subset=['Sales Margin Amount'])
```

```python
indexNames = sales[ sales['Sales Margin Amount'] <0].index
sales.drop(indexNames , inplace=True)
```

```python
df_final=grouped_df.filter(['Line Desc 1','Sales Margin Amount'], axis=1)
```
