```python
salesdetails=pd.read_csv('SALESDETAILS.csv')
region=pd.read_csv('REGION.csv')
customer=pd.read_csv('CUSTOMERS.csv')
```

```python
region.columns
```

```python
df_merge1=pd.merge(salesdetails, customer, on='Address Number')
```

```python
df_merge1.head()
```

```python
df_merge2=pd.merge(df_merge1, region, on='Region Code')
```

```python
df_merge2.head()
```

```python
df_merge2['Sales Price']=df_merge2['Sales Price'].fillna(0)
df_merge2['Sales Quantity']=df_merge2['Sales Quantity'].fillna(0)
```

```python
revenue= df_merge2['Sales Price']*df_merge2['Sales Quantity']
```

```python
df_merge2['Revenue']=revenue
```

```python
df_merge2.head(-500)
```

```python
indexNames = df_merge2[ df_merge2['Revenue'] <0].index
df_merge2.drop(indexNames , inplace=True)
```

```python
grouped_df = df_merge2.groupby(['Region Code'])
```

```python
grouped_and_summed = grouped_df.sum()
```

```python
grouped_and_summed = grouped_and_summed.reset_index()
```

```python
print(grouped_and_summed)
```

```python
df_final=grouped_and_summed.filter(['Region Code','Revenue'], axis=1)
```
