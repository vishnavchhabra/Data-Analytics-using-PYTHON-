```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salesdetails=pd.read_csv(r"C:\Users\91858\Downloads\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 -  Data\SALESDETAILS.csv")
region=pd.read_csv(r"C:\Users\91858\Downloads\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 -  Data\REGION.csv")
customer=pd.read_csv(r"C:\Users\91858\Downloads\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 Docket\Data Fiesta 2.0 Round 2 -  Data\CUSTOMERS.csv")
```

```python
region.columns
```

```python
df_merge1=pd.merge(salesdetails, customer, on='Address Number')
```

```python
df_merge1.head()
```

```python
df_merge2=pd.merge(df_merge1, region, on='Region Code')
```

```python
df_merge2.head()
```

```python
df_merge2.dropna(subset=['Sales Margin Amount'])
```

```python
indexNames = df_merge2[ df_merge2['Sales Margin Amount'] <0].index
df_merge2.drop(indexNames , inplace=True)
```

```python
grouped_df = df_merge2.groupby(['Region Code']).mean()
```

```python
df_final=grouped_df.filter(['Region Code','Sales Margin Amount'], axis=1)
```

```python
df_final
```

```python
df_final.sort_values('Sales Margin Amount',ascending=False, inplace=True)
```

```python
df_final.to_csv('marginByR.csv')
```
